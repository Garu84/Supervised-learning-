```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 500

datos = {
    'student_id': list(range(1, n+1)),                    # ID del estudiante
    'age': np.random.randint(17, 25, n),                  # Edad entre 17 y 24
    'gender': np.random.choice(['Male', 'Female'], n),    # Género
    'place_of_origin': np.random.choice(['Bogotá', 'Cundinamarca', 'Otros departamentos'], n),

    'high_school_avg': np.random.uniform(4.0, 10.0, n).round(1),      # Promedio bachillerato
    'admission_test_score': np.random.randint(150, 450, n),           # Puntaje prueba admisión
    'first_semester_grades': np.random.uniform(1.0, 5.0, n).round(1), # Promedio primer semestre

    'socioeconomic_level': np.random.randint(1, 7, n),                # Estrato 1 a 6
    'has_scholarship': np.random.choice(['Yes', 'No'], n),
    'has_loan': np.random.choice(['Yes', 'No'], n),
}

# Convert the dictionary into a DataFrame
df = pd.DataFrame(datos)

# ADDING NULL VALUES
df.loc[np.random.choice(n, 25), 'high_school_avg'] = np.nan          # 25 nulos
df.loc[np.random.choice(n, 20), 'first_semester_grades'] = np.nan
df.loc[np.random.choice(n, 15), 'admission_test_score'] = np.nan
df.loc[np.random.choice(n, 10), 'socioeconomic_level'] = np.nan

# ADDING OUTLIERS
df.loc[5, 'age'] = 35
df.loc[12, 'high_school_avg'] = 2.3
df.loc[25, 'high_school_avg'] = 10.8
df.loc[38, 'admission_test_score'] = 80
df.loc[45, 'first_semester_grades'] = 0.5
df.loc[50, 'socioeconomic_level'] = 0

# OUTPUT VARIABLE (dropout: Yes/No)
df['dropout'] = np.where(df['first_semester_grades'] < 3.0,
                         np.random.choice(['Yes', 'No'], n, p=[0.7, 0.3]),
                         np.random.choice(['Yes', 'No'], n, p=[0.2, 0.8]))

df.to_csv('synthetic_student_dropout.csv', index=False)

print("✅ ¡Dataset creado!")
print(f"   Total de filas: {len(df)}")
print(f"   Total de valores nulos: {df.isnull().sum().sum()}")
print("   Archivo guardado como: synthetic_student_dropout.csv")
```
